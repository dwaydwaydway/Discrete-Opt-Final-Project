```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
cap = pd.read_pickle(r'../cap.pck')
pref = pd.read_pickle(r'../pref.pck')
n_seats = sum(cap.values())
```

```python
capa = np.array([i for i in cap.values()])
```

```python
cost = []
percentage = np.arange(0, 0.14, 0.01)[:-1]
for i in range(14):
    with open(f'percentage_cut_0.{i:02}.pkl', 'rb') as file:
        res = pickle.load(file)
    cost.append(res[0])
```

```python
plt.plot(percentage, cost)
plt.xlabel("1-t")
plt.ylabel("Cost")
plt.show()
```

```python
cost = []
n_cut = []
for i in range(14):
    with open(f'percentage_cut_0.{i:02}.pkl', 'rb') as file:
        res = pickle.load(file)
    n_cut.append(n_seats - (capa*(1-i*0.01)).round())
    cost.append(res[0])
```

```python
plt.plot(n_cut, cost)
plt.xlabel("# of Seats Removed")
plt.ylabel("Cost")
plt.show()
```
